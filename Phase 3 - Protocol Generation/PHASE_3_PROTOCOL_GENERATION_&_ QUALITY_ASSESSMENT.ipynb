````python
# ============================================================================
# CELL 2: Define Quality Assessment Framework
# ============================================================================

print("\n" + "="*70)
print("DEFINING QUALITY ASSESSMENT FRAMEWORK")
print("="*70)

# Quality assessment criteria based on research roadmap
quality_criteria = {
    "Medical Accuracy": {
        "description": "Ensure no hallucinated drugs, doses, or procedures",
        "criteria": [
            "No hallucinated drugs or treatments",
            "Doses are within standard ranges for the disease",
            "Procedures are clinically feasible and standard of care",
            "No impossible or dangerous medical recommendations"
        ],
        "weight": 0.35,
        "scoring_guide": {
            5: "Perfect medical accuracy, all treatments standard of care",
            4: "Minor issues, easily correctable",
            3: "Moderate issues, requires review",
            2: "Major accuracy concerns, needs significant revision",
            1: "Dangerous or completely incorrect medical content"
        }
    },
    "Structural Completeness": {
        "description": "Verify all protocol sections are present and logical",
        "criteria": [
            "Title and abstract present and complete",
            "Background section covers disease context",
            "Methods section detailed with clear design",
            "Endpoints clearly defined and measurable",
            "Safety monitoring described comprehensively",
            "Statistical analysis plan included"
        ],
        "weight": 0.25,
        "scoring_guide": {
            5: "Complete protocol with all required sections",
            4: "Minor missing elements, easily added",
            3: "Some sections incomplete but usable",
            2: "Major sections missing or poorly structured",
            1: "Protocol lacks basic structure and essential sections"
        }
    },
    "Biomarker Integration": {
        "description": "Confirm biomarkers are properly integrated across systems",
        "criteria": [
            "Neurological biomarkers appropriately included",
            "Immunological biomarkers properly integrated",
            "Metabolic biomarkers relevant to disease",
            "Cross-system biomarker connections explained",
            "Biomarkers used for patient stratification"
        ],
        "weight": 0.25,
        "scoring_guide": {
            5: "Excellent cross-system biomarker integration",
            4: "Good biomarker coverage with minor gaps",
            3: "Adequate biomarker inclusion but needs improvement",
            2: "Poor biomarker integration, mostly single-system",
            1: "No meaningful biomarker integration"
        }
    },
    "Consistency & Professionalism": {
        "description": "Ensure protocols follow same format and standards",
        "criteria": [
            "Consistent format across all protocols",
            "Professional medical writing style",
            "Appropriate terminology and jargon level",
            "Logical flow and organization",
            "Free of obvious LLM artifacts or repetition"
        ],
        "weight": 0.15,
        "scoring_guide": {
            5: "Highly professional, consistent, and polished",
            4: "Professional with minor inconsistencies",
            3: "Adequate but shows some LLM characteristics",
            2: "Inconsistent format and unprofessional tone",
            1: "Poor quality, obvious LLM artifacts throughout"
        }
    }
}

# Create quality assessment template
def create_quality_assessment_template():
    """Create a template for systematic quality assessment"""

    print("Quality assessment framework defined with 4 dimensions:")
    for dimension, details in quality_criteria.items():
        print(f"   • {dimension} (Weight: {details['weight']*100:.0f}%)")
        print(f"     - {len(details['criteria'])} specific criteria")

    # Create scoring rubric DataFrame
    scoring_rubric = []
    for dimension, details in quality_criteria.items():
        for score, description in details['scoring_guide'].items():
            scoring_rubric.append({
                'Dimension': dimension,
                'Score': score,
                'Description': description
            })

    rubric_df = pd.DataFrame(scoring_rubric)

    print(f"\nScoring rubric created with {len(rubric_df)} scoring guidelines")
    print(f"Weighted scoring system ready for assessment")

    return quality_criteria, rubric_df

# Initialize the framework
quality_framework, scoring_rubric = create_quality_assessment_template()

print("\nQuality assessment framework ready!")
````

````python
# ============================================================================
# CELL 3: Quality Assessment Functions
# ============================================================================

print("\n" + "="*70)
print("CREATING QUALITY ASSESSMENT FUNCTIONS")
print("="*70)

def assess_protocol_quality(protocol_text, disease_name):
    """
    Systematically assess protocol quality across all dimensions
    Returns detailed scores and issue flags
    """

    # Initialize results dictionary
    assessment_results = {
        'medical_accuracy_score': 0,
        'structural_completeness_score': 0,
        'biomarker_integration_score': 0,
        'consistency_score': 0,
        'overall_score': 0,
        'issues_found': [],
        'corrections_needed': [],
        'ready_for_expert_review': True
    }

    # Convert protocol to lowercase for easier searching
    protocol_lower = protocol_text.lower()

    # 1. Medical Accuracy Assessment
    medical_issues = []

    # Check for common hallucination indicators
    hallucination_indicators = [
        'novel drug', 'experimental compound', 'proprietary molecule',
        'breakthrough therapy', 'miracle cure', 'revolutionary treatment',
        'never been used before', 'first-in-human', 'unprecedented results'
    ]

    for indicator in hallucination_indicators:
        if indicator in protocol_lower:
            medical_issues.append(f"Potential hallucination indicator: '{indicator}'")

    # Disease-specific medical checks
    disease_checks = {
        'Fabry Disease': {
            'required_treatments': ['enzyme replacement therapy', 'agalsidase', 'migalastat', 'chaperone therapy'],
            'forbidden_treatments': ['chemotherapy', 'radiation therapy', 'immunosuppressants']
        },
        'Dravet Syndrome': {
            'required_treatments': ['stiripentol', 'cannabidiol', 'fenfluramine', 'valproate', 'clobazam'],
            'forbidden_treatments': ['carbamazepine', 'oxcarbazepine', 'lamotrigine', 'phenytoin']
        },
        'Autoimmune Encephalitis': {
            'required_treatments': ['corticosteroids', 'ivig', 'plasmapheresis', 'rituximab', 'cyclophosphamide'],
            'forbidden_treatments': ['antibiotics alone', 'antivirals alone']
        },
        'Pediatric DIPG': {
            'required_treatments': ['radiation therapy', 'targeted therapy', 'clinical trial'],
            'forbidden_treatments': ['surgery', 'complete resection']
        },
        'Autism with Microbiome Dysbiosis': {
            'required_treatments': ['probiotics', 'prebiotics', 'fecal microbiota transplant', 'dietary intervention'],
            'forbidden_treatments': ['antipsychotics as primary treatment', 'chemotherapy']
        }
    }

    if disease_name in disease_checks:
        checks = disease_checks[disease_name]

        # Check for required treatments
        found_required = False
        for treatment in checks['required_treatments']:
            if treatment in protocol_lower:
                found_required = True
                break

        if not found_required:
            medical_issues.append(f"Missing expected standard treatment for {disease_name}")

        # Check for forbidden treatments
        for forbidden in checks['forbidden_treatments']:
            if forbidden in protocol_lower:
                medical_issues.append(f"Inappropriate treatment mentioned: '{forbidden}' for {disease_name}")

    # Score medical accuracy
    if len(medical_issues) == 0:
        assessment_results['medical_accuracy_score'] = 5
    elif len(medical_issues) <= 2:
        assessment_results['medical_accuracy_score'] = 4
    elif len(medical_issues) <= 4:
        assessment_results['medical_accuracy_score'] = 3
    elif len(medical_issues) <= 6:
        assessment_results['medical_accuracy_score'] = 2
    else:
        assessment_results['medical_accuracy_score'] = 1

    assessment_results['issues_found'].extend(medical_issues)

    # 2. Structural Completeness Assessment
    structural_issues = []
    required_sections = [
        'title', 'abstract', 'background', 'methods', 'endpoints',
        'safety monitoring', 'statistical analysis', 'references'
    ]

    missing_sections = []
    for section in required_sections:
        if section not in protocol_lower:
            missing_sections.append(section)

    if missing_sections:
        structural_issues.append(f"Missing sections: {', '.join(missing_sections)}")

    # Score structural completeness
    if len(missing_sections) == 0:
        assessment_results['structural_completeness_score'] = 5
    elif len(missing_sections) <= 2:
        assessment_results['structural_completeness_score'] = 4
    elif len(missing_sections) <= 3:
        assessment_results['structural_completeness_score'] = 3
    elif len(missing_sections) <= 4:
        assessment_results['structural_completeness_score'] = 2
    else:
        assessment_results['structural_completeness_score'] = 1

    assessment_results['issues_found'].extend(structural_issues)

    # 3. Biomarker Integration Assessment
    biomarker_issues = []

    # Check for cross-system biomarkers
    system_biomarkers = {
        'neurological': ['eeg', 'mri', 'csf', 'neurofilament', 'tau', 'amyloid'],
        'immunological': ['cytokines', 'autoantibodies', 'immune cells', 'inflammatory markers'],
        'metabolic': ['metabolites', 'enzymes', 'biomarkers', 'metabolic panels']
    }

    systems_found = 0
    for system, markers in system_biomarkers.items():
        for marker in markers:
            if marker in protocol_lower:
                systems_found += 1
                break

    if systems_found < 2:
        biomarker_issues.append(f"Limited biomarker systems: only {systems_found} system(s) covered")

    # Score biomarker integration
    if systems_found >= 3:
        assessment_results['biomarker_integration_score'] = 5
    elif systems_found == 2:
        assessment_results['biomarker_integration_score'] = 4
    elif systems_found == 1:
        assessment_results['biomarker_integration_score'] = 2
    else:
        assessment_results['biomarker_integration_score'] = 1

    assessment_results['issues_found'].extend(biomarker_issues)

    # 4. Consistency Assessment
    consistency_issues = []

    # Check for LLM artifacts
    llm_artifacts = [
        'as an ai language model', 'i am not a medical professional',
        'this is not medical advice', 'based on my training data',
        'i cannot provide medical advice', 'consult with a healthcare professional'
    ]

    for artifact in llm_artifacts:
        if artifact in protocol_lower:
            consistency_issues.append(f"LLM artifact detected: '{artifact}'")

    # Score consistency
    if len(consistency_issues) == 0:
        assessment_results['consistency_score'] = 5
    elif len(consistency_issues) <= 2:
        assessment_results['consistency_score'] = 4
    elif len(consistency_issues) <= 3:
        assessment_results['consistency_score'] = 3
    elif len(consistency_issues) <= 4:
        assessment_results['consistency_score'] = 2
    else:
        assessment_results['consistency_score'] = 1

    assessment_results['issues_found'].extend(consistency_issues)

    # Calculate overall weighted score
    weights = [
        quality_criteria['Medical Accuracy']['weight'],
        quality_criteria['Structural Completeness']['weight'],
        quality_criteria['Biomarker Integration']['weight'],
        quality_criteria['Consistency & Professionalism']['weight']
    ]

    scores = [
        assessment_results['medical_accuracy_score'],
        assessment_results['structural_completeness_score'],
        assessment_results['biomarker_integration_score'],
        assessment_results['consistency_score']
    ]

    weighted_score = sum(score * weight for score, weight in zip(scores, weights))
    assessment_results['overall_score'] = round(weighted_score, 1)

    # Determine if ready for expert review
    if assessment_results['overall_score'] < 3.0 or any('inappropriate' in issue.lower() for issue in assessment_results['issues_found']):
        assessment_results['ready_for_expert_review'] = False
        assessment_results['corrections_needed'] = [
            "Protocol requires revision before expert review",
            "Address critical medical accuracy issues",
            "Complete missing structural sections"
        ]

    return assessment_results

def create_quality_assessment_report(protocols_df):
    """
    Create comprehensive quality assessment report for all protocols
    """

    print("Starting systematic quality assessment for all protocols...")

    # Create assessment results DataFrame
    assessment_results = []

    for idx, row in protocols_df.iterrows():
        protocol_id = row['Protocol_ID']
        disease = row['Disease']
        protocol_text = row['Protocol_Text']

        print(f"   Assessing {protocol_id} ({disease})...")

        # Get quality assessment
        results = assess_protocol_quality(protocol_text, disease)

        # Add to results
        assessment_results.append({
            'Protocol_ID': protocol_id,
            'Disease': disease,
            'Medical_Accuracy_Score': results['medical_accuracy_score'],
            'Structural_Completeness_Score': results['structural_completeness_score'],
            'Biomarker_Integration_Score': results['biomarker_integration_score'],
            'Consistency_Score': results['consistency_score'],
            'Overall_Quality_Score': results['overall_score'],
            'Issues_Count': len(results['issues_found']),
            'Critical_Issues': sum(1 for issue in results['issues_found'] if 'inappropriate' in issue.lower()),
            'Ready_for_Expert_Review': results['ready_for_expert_review'],
            'Issues_Found': '; '.join(results['issues_found']) if results['issues_found'] else 'None',
            'Corrections_Needed': '; '.join(results['corrections_needed']) if results['corrections_needed'] else 'None'
        })

    # Create DataFrame
    quality_df = pd.DataFrame(assessment_results)

    print(f"\nQuality assessment completed for {len(quality_df)} protocols!")

    # Generate summary statistics
    summary_stats = {
        'total_protocols': len(quality_df),
        'protocols_ready_for_expert_review': sum(quality_df['Ready_for_Expert_Review']),
        'average_overall_score': quality_df['Overall_Quality_Score'].mean(),
        'average_medical_accuracy': quality_df['Medical_Accuracy_Score'].mean(),
        'average_structural_completeness': quality_df['Structural_Completeness_Score'].mean(),
        'average_biomarker_integration': quality_df['Biomarker_Integration_Score'].mean(),
        'average_consistency': quality_df['Consistency_Score'].mean(),
        'total_issues_found': quality_df['Issues_Count'].sum(),
        'critical_issues_found': quality_df['Critical_Issues'].sum()
    }

    return quality_df, summary_stats

print("Quality assessment functions created successfully!")
````

````python
# ============================================================================
# CELL 4: Execute Quality Assessment
# ============================================================================

print("\n" + "="*70)
print("EXECUTING SYSTEMATIC QUALITY ASSESSMENT")
print("="*70)

# Run quality assessment on all protocols
quality_df, summary_stats = create_quality_assessment_report(protocols_df)

# Merge quality assessment results with original protocols
final_df = pd.merge(protocols_df, quality_df, on=['Protocol_ID', 'Disease'])

# Save the quality assessment results
quality_csv_path = "Quality Assessment/protocol_quality_assessment.csv"
final_df.to_csv(quality_csv_path, index=False)
print(f"\nQuality assessment results saved to: {quality_csv_path}")

# Display summary statistics
print("\n" + "="*70)
print("PHASE 3 QUALITY ASSESSMENT SUMMARY")
print("="*70)
print(f"Assessment Date: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
print(f"Total Protocols Assessed: {summary_stats['total_protocols']}")
print(f"Protocols Ready for Expert Review: {summary_stats['protocols_ready_for_expert_review']}/{summary_stats['total_protocols']}")
print(f"Average Overall Quality Score: {summary_stats['average_overall_score']:.2f}/5.0")

print(f"\nDimension Scores:")
print(f"   • Medical Accuracy: {summary_stats['average_medical_accuracy']:.2f}/5.0")
print(f"   • Structural Completeness: {summary_stats['average_structural_completeness']:.2f}/5.0")
print(f"   • Biomarker Integration: {summary_stats['average_biomarker_integration']:.2f}/5.0")
print(f"   • Consistency: {summary_stats['average_consistency']:.2f}/5.0")

print(f"\nIssues Summary:")
print(f"   • Total Issues Found: {summary_stats['total_issues_found']}")
print(f"   • Critical Issues: {summary_stats['critical_issues_found']}")
print(f"   • Average Issues per Protocol: {summary_stats['total_issues_found']/summary_stats['total_protocols']:.1f}")

# Display protocols needing attention
protocols_needing_attention = final_df[~final_df['Ready_for_Expert_Review']]
if len(protocols_needing_attention) > 0:
    print(f"\nProtocols Needing Revision ({len(protocols_needing_attention)}):")
    for idx, row in protocols_needing_attention.iterrows():
        print(f"   • {row['Protocol_ID']}: Overall Score = {row['Overall_Quality_Score']:.1f}/5.0")
        print(f"     Critical Issues: {row['Critical_Issues']}")
        print(f"     Issues: {row['Issues_Found']}")
else:
    print("\nAll protocols passed quality assessment and are ready for expert review!")

print("\nQuality assessment completed successfully!")
````
